```
%matplotlib inline

#Data manipulation and linear algebra
import numpy as np
import pandas as pd

#Data Visualization
import matplotlib.pyplot as plt
import seaborn as sns

#Scaling attribute
from sklearn.preprocessing import StandardScaler

#Train-test split
from sklearn.model_selection import train_test_split

#Handling imbalanced data by oversampling the minority class
from imblearn.over_sampling import SMOTE

#Dimensionality reduction
from sklearn.decomposition import PCA

#Classifer
from sklearn.linear_model import LogisticRegression

#Evaluation metrics
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve
from sklearn.metrics import precision_recall_curve 
from sklearn.metrics import average_precision_score

#Save and load model
import pickle
```

```
#Read the csv file
df = pd.read_csv('creditcard.csv')
```

```
#View first 10 instances
df.head(10)
```

```
#View dimensions of the dataset
df.shape
```

```
#Get quick statistic summary
df.describe()
```

```
#Check the data type of all attributes
df.info()
```

```
#Check correlation
df.corr()
```

```
#See the distribution of target classes they appear to be highly imbalanced
df.Class.value_counts()
```

```
#Fraud vs. normal transactions 
counts = df.Class.value_counts()
normal = counts[0]
fraudulent = counts[1]
perc_normal = (normal/(normal+fraudulent))*100
perc_fraudulent = (fraudulent/(normal+fraudulent))*100
print('There are {} normal transactions ({:.3f}%) and {} fraudulent transactions ({:.3f}%).'.format(normal, perc_normal, fraudulent, perc_fraudulent))
```

```
#Visualize the number of transactions of each class
colors=sns.color_palette("husl", 10)
pd.Series(df['Class']).value_counts().plot(kind = "bar",color=colors,figsize=(10,8),rot=0, title = "Class")
plt.xlabel('Class', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
```

```
#Heatmap
corr = df.corr()
plt.figure(figsize=(25,25))
sns.heatmap(data=corr, annot=True, cmap="YlGnBu")
plt.title('Heatmap of Correlation')
```

```
#Create a pie chart
colors=sns.color_palette("PiYG", 10)
pd.Series(df["Class"]).value_counts().plot(kind="pie",colors=colors,labels=["Normal", "Fraudulent"],shadow=True,autopct='%.2f', fontsize=12,figsize=(6, 6),title = "Total Transactions")
```

```
#Lets drop the time attribute since it may not be as useful as other attributes
df = df.drop(columns=['Time'])
```

```
#Let us scale the amount attribute to the range 0-1
amount = df['Amount'].values.reshape(-1, 1)
scaler = StandardScaler()
scaled_array = scaler.fit_transform(amount)
amount_normalized = pd.DataFrame(scaled_array)
```

```
#View first 5 instances of the scaled attribute amount
amount_normalized.head()
```

```
#Concat the new amount to old dataframe
df = pd.concat([df, amount_normalized], axis=1)
```

```
#View the first 5 instances of the new dataframe 
df.head()
```

```
#Drop the old Amount column
df = df.drop('Amount', axis=1)
```

```
#Verify that the old amount column has been dropped
df.head()
```

```
#Specify the input attributes as X and output class as Y
X = df.drop('Class', axis=1)
Y = df['Class']
```

```
#Split the dataset into train and test sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.33, random_state=0)

print("Shape of Training features: ", X_train.shape)
print("Shape of Test features", X_test.shape)
print("Shape of Training labels", Y_train.shape)
print("Shape of Test labels", Y_test.shape)
```

```
#Oversample the minority class on the training set
sm = SMOTE(random_state=0, ratio = 1.0)
X_res, y_res = sm.fit_sample(X_train, Y_train)
```

```
#Check the new shape of resampled training input
X_res.shape
```

```
#Chcek the new shape of resampled training label
y_res.shape
```

```
#Fit Principal Component Analysis to the training input
pca = PCA()  
X_res = pca.fit_transform(X_res)  
X_test = pca.transform(X_test)  
```

```
#Observe the variance explained by each of the principal components and select the number of components that explain significant variance in the training input
explained_variance = pca.explained_variance_ratio_  
```

```
print("Explained Variance by Principal Components: ", explained_variance)
```

```
#The first Principal Component explains 67.06% of variance and second Principal Component explains 10.99%, so we will preserve 78.05% variance by selecting first two PCs
pca = PCA(n_components=2)  
X_res = pca.fit_transform(X_res)  
X_test = pca.transform(X_test)  
```

```
#Define the model and fit it to the new training data
classifier = LogisticRegression(C=1.0, random_state=0, solver='sag')
classifier.fit(X_res, y_res)
```

```
#Evaluate the model on test data
y_pred = classifier.predict(X_test)
confusion_matrix = confusion_matrix(Y_test, y_pred)
print(confusion_matrix)
```

```
#Check Precision, Recall and F1-Score of both classes (0 and 1)
print(classification_report(Y_test, y_pred))
```

```
#Out of 162 fraud transactions we correctly identify 130. We have achieved a good recall rate at the expense of bad precision. 
#Since the classes are extremely imbalanced we will prefer high recall rate.
```

```
#Check the roc_auc_score
roc_auc_score(Y_test, y_pred)
```

```
#ROC Curve
fpr, tpr, thresholds = roc_curve(Y_test, y_pred)
#Create plot
plt.plot(fpr, tpr, label='ROC curve')
plt.plot([0, 1], [0, 1], 'k--', label='Random guess')
_ = plt.xlabel('False Positive Rate')
_ = plt.ylabel('True Positive Rate')
_ = plt.title('ROC Curve')
_ = plt.xlim([-0.02, 1])
_ = plt.ylim([0, 1.02])
_ = plt.legend(loc="lower right")
```

```
#Precision vs Recall Curve
precision, recall, thresholds = precision_recall_curve(Y_test, y_pred)
#Create plot
plt.plot(precision, recall, label='Precision-recall curve')
_ = plt.xlabel('Precision')
_ = plt.ylabel('Recall')
_ = plt.title('Precision-recall curve')
_ = plt.legend(loc="lower left")
```
